```python
import numpy as np 
import pandas as pd
import os
import chardet
import matplotlib.pyplot as plt
import seaborn as sns
import spacy
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
file="spam.csv"
with open(file, 'rb') as rawdata:
    result = chardet.detect(rawdata.read(5572))
result
```

```python
data=pd.read_csv(file , encoding='ISO-8859-1')
```

```python
data
```

```python
#mapping of spam and ham into 0 and 1
data.v1.value_counts()
data.v1=pd.Categorical(data.v1).codes
data
```

```python
data.rename(columns={'v1': 'type', 'v2': 'SMS'}, inplace=True)
data=data.drop(["Unnamed: 2","Unnamed: 3","Unnamed: 4"], axis=1)
data
```

```python
data.type
```

```python
num=data.type.value_counts()
sns.barplot(x=num.index,y=num)
plt.show()
```

```python
#imbalance data management
count_class_0,count_class_1=data.type.value_counts()
```

```python
#divide it by class
data_class_0=data[data["type"]==0]
data_class_1=data[data["type"]==1]
count_class_0,count_class_1
```

```python
data_class_under=data_class_0.sample(count_class_1)
data_under=pd.concat([data_class_under,data_class_1],axis=0)
data_under.type.value_counts()
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
def preprocess(desc):
 #removing stopwords and lemmetize the text
    doc=nlp(desc)
    filtered_tokens=[]
    for token in doc:
        if token.is_stop or token.is_punct:
            continue
        filtered_tokens.append(token.lemma_)    
    return " ".join(filtered_tokens)
```

```python
data_under["New_SMS"]=data_under.SMS.apply(preprocess)
data_under
```

```python
data_under=data_under.drop("SMS",axis=1)
data_under
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
data_under.New_SMS,
data_under.type,
test_size=0.2,
random_state=1,
stratify=data_under.type    
)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
Model=Pipeline(
[
    ('vectorizer_tfidf',TfidfVectorizer()),
    ('Multi NB',MultinomialNB())
])
Model.fit(x_train,y_train)
y_pred=Model.predict(x_test)
print(classification_report(y_test,y_pred))
```
